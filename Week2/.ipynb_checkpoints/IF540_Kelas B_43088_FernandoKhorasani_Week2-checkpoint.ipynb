```python
df3 = pd.read_csv("IRIS.csv")
df3
```

```python
X = df3.iloc[:,[2,3]].values
X
```

```python
y = df3.iloc[:,4].values
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le = le.fit(y)
y = le.transform(y)
y
```

```python
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Using Min-Max Scaler to normalize data
mms = MinMaxScaler()
X_train_norm = mms.fit_transform(X_train)
X_test_norm = mms.transform(X_test)
```

```python
# X_train after normalization
X_train_norm
```

```python
# X_test after normalization
X_test_norm
```

```python
from sklearn.preprocessing import StandardScaler

# Using standard scaler to standarized data
stdsc = StandardScaler()
X_train_std = stdsc.fit_transform(X_train)
X_test_std = stdsc.transform(X_test)
```

```python
# X_train after standardization
X_train_std
```

```python
# X_test after standardization
X_test_std
```

```python
# Example of standarization and normalization (without Min-Max Scaler, etc.)
ex = np.array([0, 1, 2, 3, 4, 5])

# standarize
print('standarized:', (ex - ex.mean()) / ex.std())

# normalize
print('normalized:', (ex - ex.min()) / (ex.max() - ex.min()))
```
